```python
dfs = {}
for f in files:
    dfs[f] = akta.load_and_clean_xls(f'./data/{f}')

dfs[files[0]]['run'] = 1
dfs[files[0]] = dfs[files[0]][dfs[files[0]]['uv_280_ml'].notna()].copy()
v_max_1 = max(dfs[files[0]].uv_280_ml)

dfs[files[1]].uv_280_ml += v_max_1
dfs[files[1]].fractions_ml += v_max_1
dfs[files[1]].inject_ml += v_max_1
dfs[files[1]] = dfs[files[1]][dfs[files[1]]['uv_280_ml'].notna()].copy()
dfs[files[1]]['run'] = 2

df_hccf = pd.concat([dfs[files[0]], dfs[files[1]]], ignore_index=True)
```

```python
fig, ax = my_plot.instantiate_fig(x=20, y=6, xlabel='Volume [ml]', ylabel='A280 [mAU]')
ax.plot(df_hccf.uv_280_ml, df_hccf.uv_280_mAU)
ax.set_xlim(0, 1200)
my_plot.set_layout(fig, ax)
```

```python
v_inj_hccf = df_hccf.inject_ml[df_hccf.inject_ml.notna()].values
cycle_nums = {}
for i, v in enumerate(v_inj_hccf):
    cycle_nums[v] = i
    
cycles = []
for i, cont in df_hccf.iterrows():
    v = v_inj_hccf[v_inj_hccf - cont.uv_280_ml <= 0][-1]
    cycles.append(cycle_nums[v])
df_hccf['cycle'] = cycles

offset = 5.0
for i, v in enumerate(v_inj_hccf):
    indeces = df_hccf[df_hccf.cycle==i].index
    vals = df_hccf[df_hccf.cycle==i].uv_280_ml.values
    df_hccf.loc[indeces, 'cycle_ml'] = vals - v

for i, v in enumerate(v_inj_hccf):
    indeces = df_hccf[((df_hccf.cycle==i) & (df_hccf.cycle_ml > offset)) | ((df_hccf.cycle==i+1) & (df_hccf.cycle_ml < offset))].index
    vals = df_hccf[((df_hccf.cycle==i) & (df_hccf.cycle_ml > offset)) | ((df_hccf.cycle==i+1) & (df_hccf.cycle_ml < offset))].uv_280_ml.values
    df_hccf.loc[indeces, 'cycle_decon'] = i
    df_hccf.loc[indeces, 'cycle_decon_ml'] = vals - (v + offset)
    
```

```python
def get_mass_kDa(v):
    return np.exp(spline(v))*1e-3

def forward(v_ml):
    return np.exp(np.interp(v_ml, v_pts, np.log(m_pts)))

def inverse(m_kDa):
    return np.interp(np.log(m_kDa), np.log(m_pts[::-1]), v_pts[::-1])

def tick_function(X):
    V = forward(X)
    return ["%.0f" % z for z in V]

v_pts = np.linspace(df_std_curve['v_akta_ml'].min(), df_std_curve['v_akta_ml'].max(), 1000)
m_pts = get_mass_kDa(v_pts)
```
