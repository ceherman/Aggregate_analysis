# Load data
## SWATH

```python
df_totals = pd.read_csv('./data/elisa_bradford_and_swath_data.csv')
```

```python
dfs_swath = Dict()
dfs_swath_mab = Dict()

file = pd.ExcelFile(f'./data/Chase SWATH 20220830.xlsx')
dfs_swath.hccf.feed  = file.parse('HCCF-Feed')
dfs_swath.hccf.large = file.parse('HCCF-LrgAgg')
dfs_swath.hccf.small = file.parse('HCCF-SmAgg')
dfs_swath.hccf.mab   = file.parse('HCCF-mAb')
dfs_swath.hccf.spf1  = file.parse('HCCF-LMW1')
dfs_swath.hccf.spf2  = file.parse('HCCF-LMW2') 
dfs_swath.pavin.feed  = file.parse('PAFVIN-Feed')
dfs_swath.pavin.large = file.parse('PAFVIN-LrgAgg')
dfs_swath.pavin.small = file.parse('PAFVIN-SmAgg')

for feed in ['hccf', 'pavin']:
    for frac in dfs_swath[feed].keys():
        dfs_swath[feed][frac].columns = ['accn', 'name', 'rep1_log2_norm_area', 'rep2_log2_norm_area', 'rep3_log2_norm_area', 'prot_mw', 'rep1_ng', 'rep2_ng', 'rep3_ng', 'ave_ng', 'cv']
        dfs_swath[feed][frac]['ave_ppm'] = 10 * dfs_swath[feed][frac]['ave_ng'] # x ng/100 ug total = 10 * x ppm (in terms of total mas)
        
        # Get mAb as a separate df
        dfs_swath_mab[feed][frac] = dfs_swath[feed][frac][dfs_swath[feed][frac]['name'].str.contains('Custom')]
        dfs_swath_mab[feed][frac].reset_index(inplace=True, drop=True)
        
        # Select only CHO HCPs
        dfs_swath[feed][frac] = dfs_swath[feed][frac][dfs_swath[feed][frac]['name'].str.contains('Cricetulus griseus')]
        dfs_swath[feed][frac].reset_index(inplace=True, drop=True)
```

## DDA

```python
df_dda = pd.read_csv('./generated_tables/location_master_with_only_mean_properties.csv')
df_dda['native_hccf_spf2'] = False
df_dda.columns = ['accession', 'name', 'std_hccf_feed', 'std_hccf_large',
       'std_hccf_small', 'std_hccf_mab', 'std_hccf_spf1', 'std_hccf_spf2',
       'native_hccf_feed', 'native_hccf_large', 'native_hccf_small',
       'native_hccf_mab', 'native_hccf_spf1', 'std_pavin_feed',
       'std_pavin_large', 'std_pavin_small', 'std_pavin_mab',
       'native_pavin_feed', 'native_pavin_large', 'native_pavin_small',
       'native_pavin_mab', 'pI_mean', 'mass_mean',
       'cysteine_cont_percent_mean', 'cysteine_num_mean', 'net_charge_mean',
       'net_charge_neg_mean', 'net_charge_pos_mean', 'charge_dens_C_m2_mean',
       'charge_dens_neg_C_m2_mean', 'charge_dens_pos_C_m2_mean', 'desc_lower',
       'desc_lower_2', 'native_hccf_spf2']
```

# Process

```python
for i, cont in df_totals.iterrows():
    if cont.frac == 'spf2':
        digest = 'std'
    else:
        digest = 'native'
    df_totals.at[i, 'n_dda'] = df_dda[f'{digest}_{cont.feed}_{cont.frac}'].sum()
    df_totals.at[i, 'n_swath'] = len(dfs_swath[cont.feed][cont.frac])
```
