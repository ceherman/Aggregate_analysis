```python
# df_locs = df_master[['Standard HCCF Feed', 'Standard HCCF Large agg.', 'Standard HCCF Small agg.', 'Standard HCCF mAb', 'Standard HCCF LMW 1', 'Standard HCCF LMW 2', 
#                      'Native HCCF Feed', 'Native HCCF Large agg.', 'Native HCCF Small agg.', 'Native HCCF mAb', 'Native HCCF LMW 1',
#                      'Standard PAFVIN Feed', 'Standard PAFVIN Large agg.', 'Standard PAFVIN Small agg.', 'Standard PAFVIN mAb',
#                      'Native PAFVIN Feed', 'Native PAFVIN Large agg.', 'Native PAFVIN Small agg.', 'Native PAFVIN mAb']]

# df_locs = df_master[['Standard HCCF Feed', 'Native HCCF Feed', 'Standard HCCF Large agg.', 'Native HCCF Large agg.', 
#                      'Standard HCCF Small agg.', 'Native HCCF Small agg.', 'Standard HCCF mAb', 'Native HCCF mAb', 
#                      'Standard HCCF LMW 1', 'Native HCCF LMW 1', 'Standard HCCF LMW 2',                       
#                      'Standard PAFVIN Feed', 'Native PAFVIN Feed', 'Standard PAFVIN Large agg.', 'Native PAFVIN Large agg.', 
#                      'Standard PAFVIN Small agg.', 'Native PAFVIN Small agg.', 'Standard PAFVIN mAb', 'Native PAFVIN mAb']]

df_locs = df_master[['Native HCCF Feed', 'Standard HCCF Feed', 'Native HCCF Large agg.', 'Standard HCCF Large agg.', 
                     'Native HCCF Small agg.', 'Standard HCCF Small agg.', 'Native HCCF mAb', 'Standard HCCF mAb', 
                     'Native HCCF LMW 1', 'Standard HCCF LMW 1', 'Standard HCCF LMW 2',                       
                     'Native PAFVIN Feed', 'Standard PAFVIN Feed', 'Native PAFVIN Large agg.', 'Standard PAFVIN Large agg.', 
                     'Native PAFVIN Small agg.', 'Standard PAFVIN Small agg.', 'Native PAFVIN mAb', 'Standard PAFVIN mAb', 'name']]

for i, cont in df_locs.iterrows():
    name = cont['name']
    df_locs.at[i, 'name'] = name[:1].upper() + name[1:-21]
df_locs.set_index('name', inplace=True)

df_locs_native = df_locs[[c for c in df_locs.columns if 'Native' in c]].copy()
df_locs_std = df_locs[[c for c in df_locs.columns if 'Standard' in c]].copy()

for c in df_locs_native.columns:
    df_locs_native[c] = df_locs_native[c].map({False:0, True:1})
    
for c in df_locs_std.columns:
    df_locs_std[c] = df_locs_std[c].map({False:0, True:0.5})
    
df_locs_2 = df_locs_native.merge(df_locs_std, left_index=True, right_index=True)

df_locs_2 = df_locs_2[['Native HCCF Feed', 'Standard HCCF Feed', 'Native HCCF Large agg.', 'Standard HCCF Large agg.', 
                     'Native HCCF Small agg.', 'Standard HCCF Small agg.', 'Native HCCF mAb', 'Standard HCCF mAb', 
                     'Native HCCF LMW 1', 'Standard HCCF LMW 1', 'Standard HCCF LMW 2',                       
                     'Native PAFVIN Feed', 'Standard PAFVIN Feed', 'Native PAFVIN Large agg.', 'Standard PAFVIN Large agg.', 
                     'Native PAFVIN Small agg.', 'Standard PAFVIN Small agg.', 'Native PAFVIN mAb', 'Standard PAFVIN mAb']]
```

```python
# series = df_master['Cysteine content [%]']
series = df_master['pI']

cmap = plt.get_cmap('jet')
norm = plt.Normalize(series.min(), series.max())
    
cmap_cluster = matplotlib.colors.ListedColormap(['white', 'tab:orange', 'tab:blue'])
norm_cluster = BoundaryNorm([0, 0.25, 0.75, 1], cmap_cluster.N)

ax = sns.clustermap(df_locs_2, row_colors=cmap(norm(series)), cmap=cmap_cluster, norm=norm_cluster,
                    cbar_pos=None, dendrogram_ratio=(0.1, 0.05), figsize=(10, 15), yticklabels=False, col_cluster=False)

fontsize = 16

real_ax = ax.ax_heatmap
real_ax.set_xticklabels(real_ax.get_xmajorticklabels(), rotation=45, ha='right', rotation_mode='anchor', fontsize=fontsize)
real_ax.set_ylabel(None)

cbar_ax = ax.fig.add_axes([-0.09, 0.5, 0.02, 0.2])
cbar = matplotlib.colorbar.ColorbarBase(cbar_ax, cmap=cmap, norm=norm)
cbar.set_label(series.name, fontsize=fontsize)
cbar.ax.tick_params(labelsize=fontsize)

heat = ax.ax_heatmap
r = heat.add_patch(Rectangle((0, 0), 2, 2831, fill=False, edgecolor='black', lw=1))
r = heat.add_patch(Rectangle((2, 0), 9, 2831, fill=False, edgecolor='black', lw=1))
r = heat.add_patch(Rectangle((11, 0), 2, 2831, fill=False, edgecolor='black', lw=1))
r = heat.add_patch(Rectangle((13, 0), 6, 2831, fill=False, edgecolor='black', lw=1))

# ax.savefig(f'./images/clustermap_native_and_standard_2.png', dpi=300)
# ax.savefig(f'./images/clustermap_native_and_standard_3.png', dpi=300)
# ax.savefig(f'./images/clustermap_native_and_standard_4.png', dpi=300)
```
