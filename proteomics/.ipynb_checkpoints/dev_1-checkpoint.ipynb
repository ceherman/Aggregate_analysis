```python
def get_charge(pH, seq):
    sol = base_classes.solution(pH, ion_str=0.1)
    pro = m.protein(sol, seq=seq, dpr=2.0e-10, negative_sign=True)      
    return pro.net_charge

def check_pI(pI, seq):
    if abs(get_charge(pI, seq)) < 1e-2:
        return True
    else:
        False

def get_pI(seq, guess=5.0):
    pI = optimize.fsolve(get_charge, guess, args=(seq,))[0]
    if check_pI(pI, seq):
        return pI, True
    else:
        for guess in np.linspace(0.01, 14, 15):
            pI = optimize.fsolve(get_charge, guess, args=(seq,))[0]
            if check_pI(pI, seq):
                return pI, True
        return 0, False 
```

# Load and clean data

```python
folder = './data/'
dfs = Dict()

file = pd.ExcelFile(folder + 'CH_20220421_HCCF DDA.xlsx')
dfs.hccf.feed  = file.parse('HCCF Feed')
dfs.hccf.large = file.parse('HCCF Large agg')
dfs.hccf.small = file.parse('HCCF Small agg')
dfs.hccf.mab   = file.parse('HCCF mAb')
dfs.hccf.lmw1  = file.parse('HCCF LMW1')
dfs.hccf.lmw2  = file.parse('HCCF LMW2')

file = pd.ExcelFile(folder + 'CH_20220421_PAFVIN DDA.xlsx')
dfs.pafvin.feed  = file.parse('PAFVIN Feed')
dfs.pafvin.large = file.parse('PAFVIN Large Agg')
dfs.pafvin.small = file.parse('PAFVIN Small Agg')
dfs.pafvin.mab   = file.parse('PAFVIN mAb')
```

```python
def remove_non_CHO_and_reversed(df):
    df = df[df['Name'].str.contains('Cricetulus griseus')]
    df_rev = df[df['Accession'].str.contains('RRRRR')]
    df = df[~df['Accession'].str.contains('RRRRR')]
    df.reset_index(inplace=True, drop=True)
    return df, df_rev
```

```python
dfs_rev = Dict()

for source in dfs.keys():
    for frac in dfs[source].keys():
        dfs[source][frac], df_rev = remove_non_CHO_and_reversed(dfs[source][frac])
        if not df_rev.empty:
            df_rev.reset_index(inplace=True, drop=True)
            dfs_rev[source][frac] = df_rev
```

## Select HCPs with $\geq 2$ peptides

```python
for source in dfs.keys():
    for frac in dfs[source].keys():
        dfs[source][frac].drop(dfs[source][frac][dfs[source][frac]['Peptides(95%)'] == 1].index, inplace=True)
        dfs[source][frac].reset_index(inplace=True, drop=True)
```

## Get set of all proteins and a master dataframe with their locations

```python
names = {}
for source in dfs.keys():
    for frac in dfs[source].keys():
        for i, cont in dfs[source][frac].iterrows():
            names[cont['Accession']] = cont['Name']
            
df_master_1 = pd.DataFrame.from_dict(names, orient='index')
df_master_1.reset_index(inplace=True)
df_master_1.columns = ['accession', 'name']

for i, cont in df_master_1.iterrows():
    for source in dfs.keys():
        for frac in dfs[source].keys():
            df_master_1.at[i, f'{source}_{frac}'] = cont.accession in list(dfs[source][frac]['Accession'])

for i, cont in df_master_1.iterrows():
    text = cont['name'].replace(' [Cricetulus griseus]', '')
    df_master_1.at[i, 'desc_lower'] = text.lower()
```

### Search for repeat names

```python
hcp_names = {}
for i, n in enumerate(df_master_1.name):
    if n in hcp_names.keys():
        hcp_names[n].append(i)
    else:
        hcp_names[n] = [i]
        
dup_names = [n for n, indeces in hcp_names.items() if len(indeces) > 1]
dup_indeces = [indeces for n, indeces in hcp_names.items() if len(indeces) > 1]
dup_indeces_flat = [item for sublist in dup_indeces for item in sublist]

n_dup_max = max([len(sublist) for sublist in dup_indeces])
```

```python
unique_names = df_master_1.iloc[~df_master_1.index.isin(dup_indeces_flat)].copy()
unique_names.reset_index(inplace=True, drop=True)
unique_names.rename(columns={'accession':'accession_1'}, inplace=True)

for n in dup_names:
    df = df_master.iloc[hcp_names[n]]
```
