```python
df_risk_reviewed = pd.read_csv('./generated_tables/high_risk_reviewed.csv')

df_risk_reviewed.rename(columns={'std_hccf_feed':'Standard HCCF Feed', 'std_hccf_large':'Standard HCCF Large agg.', 'std_hccf_small':'Standard HCCF Small agg.', 
                          'std_hccf_mab':'Standard HCCF mAb', 'std_hccf_lmw1':'Standard HCCF SPF 1', 'std_hccf_lmw2':'Standard HCCF SPF 2', 
                          'native_hccf_feed':'Native HCCF Feed', 'native_hccf_large':'Native HCCF Large agg.', 'native_hccf_small':'Native HCCF Small agg.', 
                          'native_hccf_mab':'Native HCCF mAb', 'native_hccf_lmw1':'Native HCCF SPF 1', 
                          'std_pafvin_feed':'Standard PAVIN Feed', 'std_pafvin_large':'Standard PAVIN Large agg.', 'std_pafvin_small':'Standard PAVIN Small agg.', 
                          'std_pafvin_mab':'Standard PAVIN mAb', 
                          'native_pafvin_feed':'Native PAVIN Feed', 'native_pafvin_large':'Native PAVIN Large agg.', 'native_pafvin_small':'Native PAVIN Small agg.', 
                          'native_pafvin_mab':'Native PAVIN mAb',                        
                          'pI_mean':'pI', 'cysteine_cont_percent_mean':'Cysteine content [%]',
                           'cysteine_num_mean':'Cysteine residues', 'impact':'Impact'}, inplace=True)
```

```python
for i, cont in df_risk_reviewed.iterrows():
    if 'low quality protein' in cont.desc_lower:
        df_risk_reviewed.at[i, 'simple_name'] = cont.desc_lower[21:]
    else:
        df_risk_reviewed.at[i, 'simple_name'] = cont.desc_lower

df_risk_reviewed.sort_values(by=['Impact', 'simple_name'], inplace=True, ignore_index=True)
```

```python
df_locs = df_risk_reviewed[['Native HCCF Feed', 'Standard HCCF Feed', 'Native HCCF Large agg.', 'Standard HCCF Large agg.', 
                             'Native HCCF Small agg.', 'Standard HCCF Small agg.', 'Native HCCF mAb', 'Standard HCCF mAb', 
                             'Native HCCF SPF 1', 'Standard HCCF SPF 1', 'Standard HCCF SPF 2',                       
                             'Native PAVIN Feed', 'Standard PAVIN Feed', 'Native PAVIN Large agg.', 'Standard PAVIN Large agg.', 
                             'Native PAVIN Small agg.', 'Standard PAVIN Small agg.', 'Native PAVIN mAb', 'Standard PAVIN mAb',
                             'name']]

for i, cont in df_locs.iterrows():
    name = cont['name']
    df_locs.at[i, 'name'] = name[:1].upper() + name[1:-21]
df_locs.set_index('name', inplace=True)

for i, cont in df_risk_reviewed.iterrows():
    name = cont['name']
    df_risk_reviewed.at[i, 'name'] = name[:1].upper() + name[1:-21]
df_risk_reviewed.set_index('name', inplace=True)
```

```python
df_locs_native = df_locs[[c for c in df_locs.columns if 'Native' in c]].copy()
df_locs_std = df_locs[[c for c in df_locs.columns if 'Standard' in c]].copy()

for c in df_locs_native.columns:
    df_locs_native[c] = df_locs_native[c].map({False:0, True:1})
    
for c in df_locs_std.columns:
    df_locs_std[c] = df_locs_std[c].map({False:0, True:0.5})
    
df_locs_2 = df_locs_native.merge(df_locs_std, left_index=True, right_index=True)

df_locs_2 = df_locs_2[['Native HCCF Feed', 'Standard HCCF Feed', 'Native HCCF Large agg.', 'Standard HCCF Large agg.', 
                     'Native HCCF Small agg.', 'Standard HCCF Small agg.', 'Native HCCF mAb', 'Standard HCCF mAb', 
                     'Native HCCF SPF 1', 'Standard HCCF SPF 1', 'Standard HCCF SPF 2',                       
                     'Native PAVIN Feed', 'Standard PAVIN Feed', 'Native PAVIN Large agg.', 'Standard PAVIN Large agg.', 
                     'Native PAVIN Small agg.', 'Standard PAVIN Small agg.', 'Native PAVIN mAb', 'Standard PAVIN mAb']]
```

```python
cmap_impact = dict(zip(df_risk_reviewed.Impact.unique(), ["tab:green", "tab:red", "tab:purple", "tab:brown"]))
row_colors = df_risk_reviewed.Impact.map(cmap_impact)

cmap_cluster = matplotlib.colors.ListedColormap(['white', 'tab:orange', 'tab:blue'])
norm_cluster = BoundaryNorm([0, 0.25, 0.75, 1], cmap_cluster.N)

ax = sns.clustermap(df_locs_2, row_colors=row_colors, cmap=cmap_cluster, norm=norm_cluster,
                    cbar_pos=None, dendrogram_ratio=(0.1, 0.00), figsize=(18, 17), yticklabels=True, col_cluster=False, row_cluster=False)

# Add legend
handles = [Patch(facecolor=cmap_impact[name]) for name in cmap_impact]
plt.legend(handles, cmap_impact, title='Impact', bbox_to_anchor=(0.78, 0.10), bbox_transform=plt.gcf().transFigure, loc='lower right', 
           fontsize=18, frameon=False, handlelength=1, title_fontproperties={'weight':'bold', 'size':18})

# Set row color label 
rotation = 90
ax_row_colors = ax.ax_row_colors
ax_row_colors.set_xticklabels(ax_row_colors.get_xmajorticklabels(), rotation=rotation, ha='right', rotation_mode='anchor', va='center', fontsize=16, weight='bold')

# Set heatmap tick labels
heat = ax.ax_heatmap
heat.set_xticklabels(heat.get_xmajorticklabels(), rotation=rotation, ha='right', rotation_mode='anchor', va='center', fontsize=16)
heat.set_yticklabels(heat.get_ymajorticklabels(), fontsize=14)
heat.set_ylabel(None)

# Add rectangles for grouping the data
cnt = 0
for n in [1, 1, 1, 1, 4, 4, 2, 5, 4, 4, 1, 1, 14, 5, 5, 1, 2]:
    heat.add_patch(Rectangle((0, cnt), 19, n, fill=False, edgecolor='black', lw=1))
    cnt += n
heat.add_patch(Rectangle((0, 0), 2, len(df_locs_2), fill=False, edgecolor='black', lw=1))
heat.add_patch(Rectangle((11, 0), 2, len(df_locs_2), fill=False, edgecolor='black', lw=1))

ax.savefig(f'./images/5_heatmap_high_risk_native_and_standard.png', dpi=300)
ax.savefig(f'./images/5_heatmap_high_risk_native_and_standard.tif', dpi=300)
ax.savefig(f'./images/5_heatmap_high_risk_native_and_standard.eps', dpi=300)
```
